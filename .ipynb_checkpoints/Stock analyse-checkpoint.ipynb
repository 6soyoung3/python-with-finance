```python
import pandas as pd
#pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import matplotlib.dates as md
import math
```

```python
spydata = pd.read_csv('spy.csv')
print(spydata.columns) # date, open, high, low, close, adj close, volume
spydata.head()
```

```python
spydata['Date'] = pd.to_datetime(spydata['Date'])
spydata.insert(0, 'Year', spydata['Date'].dt.year)
spydata.insert(1, 'Month', spydata['Date'].dt.month)
spydata.insert(2, 'Day', spydata['Date'].dt.day)
spydata['Date'] = spydata['Date'].dt.date

spydata.head()
```

```python
fig, ax = plt.subplots(4, 2, figsize=(16,10))

data = spydata.melt(id_vars=['Date','Year'], value_vars=['Open','Close'], var_name = 'State', value_name='Price')

data1993 = data.loc[data['Year'] == 1993]
data1994 = data.loc[data['Year'] == 1994]
data1995 = data.loc[data['Year'] == 1995]
data1996 = data.loc[data['Year'] == 1996]
data1997 = data.loc[data['Year'] == 1997]
data1998 = data.loc[data['Year'] == 1998]
data1999 = data.loc[data['Year'] == 1999]

# data90sList = [data1993, data1994, data1995, data1996, data1997, data1998, data1999]

fig.suptitle('Overall opening and closing prices across 90s')

sns.lineplot(ax=ax[0,0], x=data1993.Date, y= data1993.Price, hue= data1993.State, legend=False)
sns.lineplot(ax=ax[0,1], x=data1994.Date, y= data1994.Price, hue= data1994.State, legend=False)
sns.lineplot(ax=ax[1,0], x=data1995.Date, y= data1995.Price, hue= data1995.State, legend=False)
sns.lineplot(ax=ax[1,1], x=data1996.Date, y= data1996.Price, hue= data1996.State, legend=False)
sns.lineplot(ax=ax[2,0], x=data1997.Date, y= data1997.Price, hue= data1997.State, legend=False)
sns.lineplot(ax=ax[2,1], x=data1998.Date, y= data1998.Price, hue= data1998.State, legend=False)
sns.lineplot(ax=ax[3,0], x=data1999.Date, y= data1999.Price, hue= data1999.State, legend=False)

fig.legend(['Open','Close'])

plt.tight_layout()
plt.show()
```

- Graphs above shows how opening and closing prices have grown across 90s ('93 to '99)
- Overall, SPY's opening and closing prices were similar which means within single day, there was no big gap between opening and closing prices but we can look closer by graph below
- Out of 7, 6 of them shows gentle grow like y = x except '94. Between 94-03 and 94-05, 

```python
data90s = spydata.loc[spydata['Year'] < 2000]
allGapOpenClose = []
for index, row in data90s.iterrows():
    allGapOpenClose.append([row['Year'],row['Date'], row['Close']-row['Open']])
    
dfOpenClose = pd.DataFrame(allGapOpenClose,columns=['Year','Date','Close - Open'])

data1993 = (dfOpenClose.loc[dfOpenClose['Year'] == 1993]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]
data1994 = (dfOpenClose.loc[dfOpenClose['Year'] == 1994]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]
data1995 = (dfOpenClose.loc[dfOpenClose['Year'] == 1995]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]
data1996 = (dfOpenClose.loc[dfOpenClose['Year'] == 1996]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]
data1997 = (dfOpenClose.loc[dfOpenClose['Year'] == 1997]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]
data1998 = (dfOpenClose.loc[dfOpenClose['Year'] == 1998]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]
data1999 = (dfOpenClose.loc[dfOpenClose['Year'] == 1999]).sort_values(by=['Close - Open'], axis=0).iloc[0:5]

fig, axes = plt.subplots(4, 2, figsize=(16,10))

fig.suptitle('Gaps between closing and opening prices across 90s (Decreased)')

sns.barplot(ax=axes[0,0], x=data1993.Date, y= data1993['Close - Open'])
sns.barplot(ax=axes[0,1], x=data1994.Date, y= data1994['Close - Open'])
sns.barplot(ax=axes[1,0], x=data1995.Date, y= data1995['Close - Open'])
sns.barplot(ax=axes[1,1], x=data1996.Date, y= data1996['Close - Open'])
sns.barplot(ax=axes[2,0], x=data1997.Date, y= data1997['Close - Open'])
sns.barplot(ax=axes[2,1], x=data1998.Date, y= data1998['Close - Open'])
sns.barplot(ax=axes[3,0], x=data1999.Date, y= data1999['Close - Open'])

plt.tight_layout()
plt.show()
```

```python
data1993 = (dfOpenClose.loc[dfOpenClose['Year'] == 1993]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]
data1994 = (dfOpenClose.loc[dfOpenClose['Year'] == 1994]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]
data1995 = (dfOpenClose.loc[dfOpenClose['Year'] == 1995]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]
data1996 = (dfOpenClose.loc[dfOpenClose['Year'] == 1996]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]
data1997 = (dfOpenClose.loc[dfOpenClose['Year'] == 1997]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]
data1998 = (dfOpenClose.loc[dfOpenClose['Year'] == 1998]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]
data1999 = (dfOpenClose.loc[dfOpenClose['Year'] == 1999]).sort_values(by=['Close - Open'], axis=0, ascending=False).iloc[0:5]

fig, axes = plt.subplots(4, 2, figsize=(16,10))

fig.suptitle('Gaps between closing and opening prices across 90s (Increased)')

sns.barplot(ax=axes[0,0], x=data1993.Date, y= data1993['Close - Open'])
sns.barplot(ax=axes[0,1], x=data1994.Date, y= data1994['Close - Open'])
sns.barplot(ax=axes[1,0], x=data1995.Date, y= data1995['Close - Open'])
sns.barplot(ax=axes[1,1], x=data1996.Date, y= data1996['Close - Open'])
sns.barplot(ax=axes[2,0], x=data1997.Date, y= data1997['Close - Open'])
sns.barplot(ax=axes[2,1], x=data1998.Date, y= data1998['Close - Open'])
sns.barplot(ax=axes[3,0], x=data1999.Date, y= data1999['Close - Open'])

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(16,28))

fig.suptitle('Overall opening and closing prices across decades')

data = spydata.melt(id_vars=['Date','Year', 'Adj Close'], value_vars=['Open','Close'], var_name = 'State', value_name='Price')
data = data.drop(data[data.Year == 2020].index)

col = ['Average adj close']
lst = []
years = list(range(1993, 2020))

for x in range(1993, 2020):
    lst.append(spydata.loc[spydata['Year'] == x]['Adj Close'].mean())

eachYearMean = pd.DataFrame(lst, columns=col)
eachYearMean['Year'] = years

sns.boxplot(x = data.Year, y = data.Price, data = data, hue = data.State, ax=ax1)
sns.lineplot(data=eachYearMean['Average adj close'], label="Average adjusted closing price", ax=ax1)

plt.tight_layout()
plt.show()
```
